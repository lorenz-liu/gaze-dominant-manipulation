```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = ['Arial Unicode MS']

#  x 
x = np.linspace(-1, 1, 100)

#  y = x 
y1 = x

#  y = 5x^5 
y2 = 5 * x ** 5

# 
fig, ax = plt.subplots()
ax.plot(x, y1, label='Original Gaze')
ax.plot(x, y2, label='Adapted Gaze')

#  x  y 
ax.set_xlim(-1, 1)
ax.set_ylim(-5, 5)

#  x  y  0.2
ax.set_xticks(np.arange(-1, 1.2, 0.2))
ax.set_yticks(np.arange(-5, 5.2, 1))

# 
ax.grid()

# 
ax.legend()

#  x  y 
ax.set_xlabel('v\n(the original offset of gaze)', fontsize=16)
ax.set_ylabel('(the adapted value for manipulation)\nA', fontsize=16)

# 
plt.show()
```
